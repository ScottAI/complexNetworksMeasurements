```python
%matplotlib.pyplot as pylab inline
import networkx as nx
```

```python
def koch_network(m, t):
    # Create initial graph
    g = nx.Graph()
    g.add_edge(1, 2)
    g.add_edge(1, 3)
    g.add_edge(2, 3)

    new_node = 4

    for step in range(t):
        for node in g.nodes():
            for i in range(nx.triangles(g, node)):
                for group in range(m):
                    for j in range(2):
                        g.add_node(new_node)
                        g.add_edge(node, new_node)

                        # Add the edge between the two new nodes
                        if j is 1:
                            g.add_edge(new_node, new_node-1)

                        new_node += 1

    return g
```

```python
koch_1_2 = koch_network(1, 2)
# print("triangles: ", nx.triangles(koch_1_2))
print("edges: ", koch_1_2.edges())
nx.draw_spring(koch_1_2)
```
